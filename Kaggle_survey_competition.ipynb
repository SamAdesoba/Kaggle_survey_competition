# COUNTRIES ANNALYSIS

```python
df = df[~df['Q4'].isin(['Other', 'In which country do you currently reside?', 'I do not wish to disclose my location'])]
df.head()
```

```python
df['Q4'] = df['Q4'].replace({'Hong Kong (S.A.R.)':'Hong Kong','Iran, Islamic Republic of...': 'Iran'})
```

```python
df_country = df.Q4
df_country.value_counts().head()
```

```python
def convert(row):
    country_code = pc.country_name_to_country_alpha2(row.Q4, cn_name_format= "default")
    conti_code = pc.country_alpha2_to_continent_code(country_code)
    return conti_code
```

```python
df['continent'] = df.apply(convert, axis = 1)
df.head()
```

```python
conti_names = {
    "AS": "Asia",
    "SA": "South America",
    "OC": "Australia",
    "EU": "Europe",
    "NA": "North America",
    "AF": "Africa"
}
df['continent'] = df['continent'].map(conti_names)
df.head()
```

```python
df_continent = df.continent
df_continent.head()
```

```python
continent_sum = df_continent.value_counts()
continent_sum
```

```python
width = 0.35
plt.figure(figsize=(10, 5))
plt.figure(figsize=(6,6), dpi=150)
plt.subplot(111)
plt.title('Variation with Data Scientist in continents')
plt.ylabel("Continents")
continent_sum.plot(kind = 'pie', autopct='%1.1f%%', fontsize=12)
plt.show()
```
